# Example Gamma-ray source analysis

This notebook details the data reduction in which observations of the High Energy Stereoscopic System (H.E.S.S.) are stacked into a single dataset.

Quality cuts are performed to discard data that is not optimal, and 2 different background estimation methods are shown.

Then, a 3-dimensional source analysis is performed, deriving the morphology and the spectrum of the Crab Nebula.

```python
data = {
    'x_center': [],
    'x_err': [],
    'y_center': [],
    'err_high': [],
    'err_low': []
}

# Populate the DataFrame
for i in range(len(flux_points_001.e2dnde.data)):
    if not flux_points_001.is_ul.data[i][0][0]:
        x_center = flux_points_001.energy_ref[i].value
        x_err = flux_points_001.energy_max[i].value - x_center
        y_center = flux_points_001.e2dnde.data[i][0][0]
        err_high = flux_points_005.e2dnde.data[i][0][0] + flux_points_005.e2dnde_errp.data[i][0][0] - y_center
        err_low = y_center - (flux_points_004.e2dnde.data[i][0][0] - flux_points_004.e2dnde_errn.data[i][0][0])
        
        data['x_center'].append(x_center)
        data['x_err'].append(x_err)
        data['y_center'].append(y_center)
        data['err_high'].append(err_high)
        data['err_low'].append(err_low)

# Convert the data to a DataFrame
df = pd.DataFrame(data)
```

```python
data = {
    'energy_center': [],
    'energy_error': [],
    'res_rm': [],
    'res_old': [],
    'error_rm': [],
    'error_low': [],
    'error_high': [],
    'error_old_low': [],
    'error_old_high': []
}

# Populate the DataFrame
for i in range(len(df['y_center'])):
    for j in range(len(df_old['flux_old'])):
        if not flux_points_001.is_ul.data[i][0][0] and \
           df_old['e'].iloc[j] - 0.1 < flux_points_001.energy_ref.data[i] < df_old['e'].iloc[j] + 0.1:
            x1 = df_old['flux_old'].iloc[j]
            x2 = flux_points_001.e2dnde.data[i][0][0]
            x_ref = sky_model_01.spectral_model.reference_fluxes(energy_axis)['ref_e2dnde'][i]
            value1 = (x1 - x_ref.value) / x_ref.value
            value2 = (x2 - x_ref.value) / x_ref.value
            
            delx2 = flux_points_001.e2dnde_err.data[i][0][0]
            delx3 = df['err_low'].iloc[i]
            delx4 = df['err_high'].iloc[i]

            data['energy_center'].append(df_old['e'].iloc[j])
            data['energy_error'].append((flux_points_001.energy_max.data[i] - flux_points_001.energy_min.data[i]) / 2)
            data['res_rm'].append(value2)
            data['res_old'].append(value1)
            data['error_old_low'].append(df_old['flux_old_min'].iloc[j] / x_ref.value)
            data['error_old_high'].append(df_old['flux_old_max'].iloc[j] / x_ref.value)
            data['error_rm'].append(delx2 / x_ref.value)
            data['error_low'].append(delx3 / x_ref.value)
            data['error_high'].append(delx4 / x_ref.value)

# Convert the data to a DataFrame
df_res = pd.DataFrame(data)
```
