## Zenith vs trigger rate

```python
def fit_func(x,a,b):
    return a * (np.cos(np.deg2rad(x)))**b
```

```python
def fit_func2(x, a, b, c, d):
    cos = 1-np.cos(np.deg2rad(x))
    numerator = a - b*cos
    denominator = np.exp((x-c)/d)+1
    return numerator/denominator
```

```python
def fit_func3(x, a, c, d):
    cos = 1-np.cos(np.deg2rad(x))
    numerator = a - 0*cos
    denominator = np.exp((x-c)/d)+1
    return numerator/denominator
```
