```python
def calculate_mean(intervals, dataset): 
    ''' 
    Group the input into bins according to their fractional run deviation. Get the mean and standard deviation in each bin

    Returns pd.Dataframe filled with this information

    ---------------------------------------------------------------------------
    Parameters:
    
    - interevals: list, fractional run deviaton bins

    - dataset: pd.Dataframe, filled with the frac run deviation and bkg deviation for all run pairs that should be analised
    
    '''
    
    all_set = pd.DataFrame(columns=['run dev', 'run dev err', 'sys dev', 'sys dev error'])
    
    for i in range(0,len(intervals)-1):
        dataframe = dataset[(dataset['deviation'].astype(float)>intervals[i]) & (dataset['deviation'].astype(float)<intervals[i+1])]  

        #for low statistics no sensible values can be calculated, therefore calculate std only for frac run dev with more than 10 entries
        if len(dataframe) >= 10:
            background = dataframe['bkg_dev']
            background_error = dataframe['bkg_dev error']
        
            if not background.empty:
                mu, std = norm.fit(background)
                systematic_data = [(intervals[i] + intervals[i+1])/2, (intervals[i+1] - intervals[i])/2, mu, std]
                new_df = pd.DataFrame( [systematic_data], 
                                         columns=['run dev', 'run dev err', 'sys dev', 'sys dev error'])
                all_set = pd.concat([all_set.astype(new_df.dtypes), new_df], ignore_index=True)
            
    return all_set
```

```python
all_set1 = calculate_mean(intervals=np.linspace(0,3,31), dataset=dataset1)
all_set2 = calculate_mean(intervals=np.linspace(0,3,31), dataset=dataset2)
all_set3 = calculate_mean(intervals=np.linspace(0,3,31), dataset=dataset3)
```

```python
out_path = '/home/wecapstor1/caph/mppi103h/On-Off-matching/systematic_error_estimation'
all_set1.iloc[:-3].to_csv('{}/systematic_errors_dataset1.csv'.format(out_path), sep='\t')
all_set2.iloc[:-1].to_csv('{}/systematic_errors_dataset2.csv'.format(out_path), sep='\t')
all_set3.iloc[:-2].to_csv('{}/systematic_errors_dataset3.csv'.format(out_path), sep='\t')
```
