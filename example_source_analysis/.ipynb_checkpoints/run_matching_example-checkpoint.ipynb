```python
from astropy.coordinates import SkyCoord
from astropy import units as u

# Query the database for all runs which can be used as OFF runs, the declination and RunType cut help to filter out calibration runs
def query_off_runs():
    run_list = []
    for i in range(0,len(monitor_run_data_use)):
        #print(monitor_run_data['Run'][i])
        ra = monitor_run_data_use['Target_RA'].iloc[i]
        dec = monitor_run_data_use['Target_Dec'].iloc[i]
        name = monitor_run_data_use['Target_Name'].iloc[i]
        if monitor_run_data['RunType'].iloc[i] == 'ObservationRun':
            if dec > -90 and dec < 90:
                coord = SkyCoord(ra=ra*u.degree, dec=dec*u.degree, frame='icrs').galactic
                if coord.b.value > 10 or coord.b.value < -10:
                    #if name[0:3]!='ToO' and name[0:3]!='GRB' and name not in targets_with_detection:
                    if name not in targets_with_detection:
                        run_list.append(monitor_run_data_use['Run'].iloc[i])
    
    return run_list
```

```python
def checkHessEra(HessEra):
    if (HessEra != "HESS1") and (HessEra != "HESS2"):
        print("ERROR! This HessEra '"+HessEra+"' doesn't exist. Please choose between 'HESS1' or 'HESS2'")
        raise
```

```python
def checkConfig(config):
    fHybrid, fStereo, fMono = False, False, False
    if config == "hybrid":
        fHybrid = True
    elif config == "stereo":
        fStereo = True
    elif config == "mono":
        fMono = True
    else:
        print("ERROR! Unknown config '"+config+"'. Please choose between 'hybrid','stereo' or 'mono'")

    return fHybrid, fStereo, fMono
```

```python
def quality_cuts(run_list,config="stereo", analysis="",HessEra="", requireCT5=False, logBadRuns=False):
    """
    Apply quality selection cuts on your runlist
    --------------------------------------------
    Parameters:
        - run_list: numpy.Array

        - config: string

        - analysis: string

        - HessEra:

        - requireCT5: bool

        - logBadRuns: bool
    """

    # First some safety checks (spelling errors etc.)
    checkHessEra(HessEra)
    fHybrid, fStereo, fMono = checkConfig(config)

    # Deciding between different analysis types
    fSpectral, fDetection = False, False
    if analysis == "spectral":
        fSpectral = True
        HESS_Quality_Criteria_file = "/home/woody/caph/mppi103h/runselection/quality_run_selection/"+HessEra+"_SPECTRAL_CRITERIA.yaml"
    elif analysis == "detection":
        fDetection = True
        HESS_Quality_Criteria_file ="/home/woody/caph/mppi103h/runselection/quality_run_selection/"+HessEra+"_DETECTION_CRITERIA.yaml"
    else:
        print("WRONG! Choose between 'detection' or 'spectral'.")
        raise

    # Loading in the different Quality selection criteria
    HESS_Quality_Criteria =  open(HESS_Quality_Criteria_file,'r')
    HESS_Quality_Criteria = yaml.safe_load(HESS_Quality_Criteria)
    runs_failed = []
    runlist_selected = []

    if(logBadRuns):
        logFile = open("logBadQualityRuns.out", "w")
        logFile.write("Here you can find the 'bad quality' runs --> always suggested to check the websummary \n")
        logFile.close()


    
    for run in run_list:
        n_failed_criteria = 0
        failed_criteria_logs = []
        
        
        #livetime
        tested_run = monitor_run_data[monitor_run_data['Run']==run]
        if tested_run.empty == True:
            if monitor_run_data[monitor_run_data['Run']==run].empty == False:
                if monitor_run_data[monitor_run_data['Run']==run]['Duration'].iloc[0]!= 0:
                    if monitor_run_data[monitor_run_data['Run']==run]['Duration'].iloc[0]<HESS_Quality_Criteria["Duration"][0] or DB_analysis[DB_analysis['Run']==run]['Duration'].iloc[0]>HESS_Quality_Criteria["Duration"][1] :
                        runs_failed.append(run)
                        n_failed_criteria += 1
                        if(logBadRuns):
                            failed_criteria_logs.append("   - Duration criteria [{},{}] failed: run duration {} ".format(HESS_Quality_Criteria["Duration"][0],HESS_Quality_Criteria["Duration"][1],tested_run['Duration'].iloc[0]))

        if tested_run.empty == False:
                if tested_run['Duration'].iloc[0]<HESS_Quality_Criteria["Duration"][0] or tested_run['Duration'].iloc[0]>HESS_Quality_Criteria["Duration"][1] :
                    runs_failed.append(run)
                    n_failed_criteria += 1
                    if(logBadRuns):
                        failed_criteria_logs.append("   - Duration criteria [{},{}] failed: run duration {} ".format(HESS_Quality_Criteria["Duration"][0],HESS_Quality_Criteria["Duration"][1],tested_run['Duration'].iloc[0]))


        #participation frac
        tested_run = DB_header[DB_header['Run']==run]
        if tested_run.empty ==True:
            runs_failed.append(run)
            n_failed_criteria += 1
            #failed_criteria_logs.append("   - Missing data {Participation_frac} ")

        if tested_run.empty == False:
            if(fMono):
                tested_run = tested_run[tested_run["Telescope"] == 5]
            if((fStereo==True) & (requireCT5!=True)):
                tested_run = tested_run[tested_run["Telescope"] != 5]

            for tel_id in np.unique(tested_run["Telescope"]):
                tested_run_telid = tested_run[tested_run["Telescope"]==tel_id]
                if config=='mono' or config=='hyrbid':
                    if tested_run_telid['Participation_frac'].iloc[0]<HESS_Quality_Criteria["Participation_frac_"+config][0] or tested_run_telid['Participation_frac'].iloc[0]>HESS_Quality_Criteria["Participation_frac_"+config][1]:
                        runs_failed.append(run)
                        n_failed_criteria += 1
                        if(logBadRuns):
                            failed_criteria_logs.append("   - Participation fraction criteria [{},{}] failed for CT{} : run Participation fraction of CT{} {} ".format(HESS_Quality_Criteria["Participation_frac_"+config][0],HESS_Quality_Criteria["Participation_frac_"+config][1],tel_id,tel_id,tested_run_telid['Participation_frac'].iloc[0]))
                if config=='stereo':
                    if monitor_run_data[monitor_run_data['Run']==run]['Telescope_Pattern'].iloc[0] == 30:
                        if tested_run_telid['Participation_frac'].iloc[0]<HESS_Quality_Criteria["Participation_frac_"+config][0] or tested_run_telid['Participation_frac'].iloc[0]>HESS_Quality_Criteria["Participation_frac_"+config][1]:
                            runs_failed.append(run)
                            n_failed_criteria += 1
                            if(logBadRuns):
                                failed_criteria_logs.append("   - Participation fraction criteria [{},{}] failed for CT{} : run Participation fraction of CT{} {} ".format(HESS_Quality_Criteria["Participation_frac_"+config][0],HESS_Quality_Criteria["Participation_frac_"+config][1],tel_id,tel_id,tested_run_telid['Participation_frac'].iloc[0]))
                    if monitor_run_data[monitor_run_data['Run']==run]['Telescope_Pattern'].iloc[0] == 62:
                            if tested_run_telid['Participation_frac'].iloc[0]<HESS_Quality_Criteria["Participation_frac_hybrid"][0] or tested_run_telid['Participation_frac'].iloc[0]>HESS_Quality_Criteria["Participation_frac_hybrid"][1]:
                                runs_failed.append(run)
                                n_failed_criteria += 1
                                if(logBadRuns):
                                    failed_criteria_logs.append("   - Participation fraction criteria [{},{}] failed for CT{} : run Participation fraction of CT{} {} ".format(HESS_Quality_Criteria["Participation_frac_hybrid"][0],HESS_Quality_Criteria["Participation_frac_hybrid"][1],tel_id,tel_id,tested_run_telid['Participation_frac'].iloc[0]))




        #tracking criteria
        tested_run = DB_track[DB_track['Run']==run]
        if(fMono):
            tested_run = tested_run[tested_run["Telescope"] == 5]
        if((fStereo==True) & (requireCT5!=True)):
            tested_run = tested_run[tested_run["Telescope"] != 5]

        for tel_id in np.unique(tested_run["Telescope"]):
            if tested_run.empty ==True:
                runs_failed.append(run)
                n_failed_criteria += 1
                #failed_criteria_logs.append("   - Missing data {Tracking} for CT{} ".format(tel_id))
            if tested_run.empty == False:
                e1 = tested_run[tested_run["Telescope"]==tel_id]["Alt_Dev_rms"].iloc[0]
                e2 = tested_run[tested_run["Telescope"]==tel_id]["Az_Dev_rms"].iloc[0]
                e3 = tested_run[tested_run["Telescope"]==tel_id]["RA_Dev_mean"].iloc[0]
                e4 = tested_run[tested_run["Telescope"]==tel_id]["Dec_Dev_mean"].iloc[0]

                if e1<HESS_Quality_Criteria["Alt_Dev_rms"][0] or e1>HESS_Quality_Criteria["Alt_Dev_rms"][1] or e2<HESS_Quality_Criteria["Alt_Dev_rms"][0] or e2>HESS_Quality_Criteria["Alt_Dev_rms"][1] or e3<HESS_Quality_Criteria["RA_Dev_mean"][0] or e3>HESS_Quality_Criteria["RA_Dev_mean"][1] or e4<HESS_Quality_Criteria["Dec_Dev_mean"][0] or e4>HESS_Quality_Criteria["Dec_Dev_mean"][1]:
                    runs_failed.append(run)
                    n_failed_criteria += 1
                    if(logBadRuns):
                        failed_criteria_logs.append("   - Tracking criteria failed for CT{}".format(tel_id))


        #broken pixel
        #DB_broken_pix = self.DB_pix[(self.DB_pix['Livetime']>0)].sort_values("WhenEntered", ascending=False).drop_duplicates(subset=['Run'], keep='last')
        tested_run = DB_pix[DB_pix['Run']==run]
        if tested_run.empty ==True:
            runs_failed.append(run)
            n_failed_criteria += 1
            #failed_criteria_logs.append("   - Missing data {Broken pixel}")
        if tested_run.empty == False:
            if(fMono):
                tested_run = tested_run[tested_run["Telescope"] == 5]
            if((fStereo==True) & (requireCT5!=True)):
                tested_run = tested_run[tested_run["Telescope"] != 5]

            for tel_id in np.unique(tested_run["Telescope"]):
                tested_run_telid = tested_run[tested_run["Telescope"]==tel_id]
                if tested_run_telid['Num_Hardware'].iloc[0]<HESS_Quality_Criteria["Num_Hardware_"+config][0] or tested_run_telid['Num_Hardware'].iloc[0]>HESS_Quality_Criteria["Num_Hardware_"+config][1] or tested_run_telid['Num_HV_Turned_Off'].iloc[0]<HESS_Quality_Criteria["Num_HV_Turned_Off_"+config][0] or tested_run_telid['Num_HV_Turned_Off'].iloc[0]>HESS_Quality_Criteria["Num_HV_Turned_Off_"+config][1]:
                    runs_failed.append(run)
                    n_failed_criteria += 1
                    if(logBadRuns):
                        failed_criteria_logs.append("   - Broken pixel criteria failed for CT{} ".format(tel_id))


        #trigger criteria
#         print(run)
        tested_run = DB_trigger[DB_trigger['Run']==run]
        if(fMono):
            tested_run = tested_run[tested_run["Telescope"] == 5]
            track_tel = DB_track[DB_track['Run']==run][DB_track["Telescope"]== 5]
        if((fStereo==True) & (requireCT5!=True)):
            tested_run = tested_run[tested_run["Telescope"] != 5]
            track_tel = DB_track[DB_track['Run']==run][DB_track["Telescope"]!= 5]
        if tested_run.empty ==True:
            runs_failed.append(run)
            n_failed_criteria += 1
            failed_criteria_logs.append("   - Missing data {Trigger rate}")
        if tested_run.empty ==False:
            for tel_id in np.unique(track_tel['Telescope']):
                if tested_run[tested_run["Telescope"]==tel_id].empty == False:
                    tested_run_telid = tested_run[tested_run["Telescope"]==tel_id]
                    if HessEra== 'HESS2':
                        if(fSpectral):
                            if tested_run_telid['True_Rate_Delta_1'].iloc[0]<HESS_Quality_Criteria["True_Rate_Delta_1"][0] or tested_run_telid['True_Rate_Delta_1'].iloc[0]>HESS_Quality_Criteria["True_Rate_Delta_1"][1] or tested_run_telid['True_Rate_Delta_2'].iloc[0]<HESS_Quality_Criteria["True_Rate_Delta_2"][0] or tested_run_telid['True_Rate_Delta_2'].iloc[0]>HESS_Quality_Criteria["True_Rate_Delta_2"][1]:
                                runs_failed.append(run)
                                n_failed_criteria += 1
                                if(logBadRuns):
                                    failed_criteria_logs.append("   - Trigger rate criteria failed for CT{}".format(tel_id))
                    if HessEra== 'HESS1':
                        if(fSpectral):
                            #print(tested_run_telid)
                            if tested_run_telid['True_Rate_Delta_1'].iloc[0]<HESS_Quality_Criteria["True_Rate_Delta_1"][0] or tested_run_telid['True_Rate_Delta_1'].iloc[0]>HESS_Quality_Criteria["True_Rate_Delta_1"][1] or tested_run_telid['True_Rate_Delta_2'].iloc[0]<HESS_Quality_Criteria["True_Rate_Delta_2"][0] or tested_run_telid['True_Rate_Delta_2'].iloc[0]>HESS_Quality_Criteria["True_Rate_Delta_2"][1]:
                                runs_failed.append(run)
                                n_failed_criteria += 1
                                if(logBadRuns):
                                    failed_criteria_logs.append("   - Trigger rate criteria failed for CT{}".format(tel_id))
                if tested_run[tested_run["Telescope"]==tel_id].empty == True: 
                    runs_failed.append(run)
                    n_failed_criteria += 1
                    failed_criteria_logs.append("   - Missing data {Trigger rate}")
                        
        #atmospheric criteria
        if(fSpectral):
            tested_run = DB_atmo[DB_atmo['Run']==run]
            if tested_run.empty ==True:
                runs_failed.append(run)
                n_failed_criteria += 1
                #failed_criteria_logs.append("   - Missing data {Transparency Coefficient}")
            if DB_atmo[DB_atmo['Run']==run].empty == False:
                if tested_run['TransparencyCoefficient_mean'].iloc[0] != 0:
                    if tested_run['TransparencyCoefficient_mean'].iloc[0]<HESS_Quality_Criteria['TransparencyCoefficient_mean'][0] or tested_run['TransparencyCoefficient_mean'].iloc[0]>HESS_Quality_Criteria['TransparencyCoefficient_mean'][1]:
                        runs_failed.append(run)
                        n_failed_criteria += 1
                        if(logBadRuns):
                            failed_criteria_logs.append("   - Atmospheric rate criteria failed")

        if logBadRuns and n_failed_criteria > 0:
            os.system("echo Run {} failed {} quality criteria: >> logBadQualityRuns.out".format(run,n_failed_criteria))
            os.system("echo ---------------------------------- >> logBadQualityRuns.out")

            if len(failed_criteria_logs) == 0:
                os.system("echo MISSING DATA >> logBadQualityRuns.out")
            else:
                for i in range(len(failed_criteria_logs)):
                    os.system("echo "+failed_criteria_logs[i]+">> logBadQualityRuns.out")
            os.system("echo   >> logBadQualityRuns.out")


    runlist_selected_runs = [x for x in run_list if x not in runs_failed]
    for run in runlist_selected_runs:
        runlist_selected.append(run) 

    return runlist_selected
```

```python
def filter_era(era, off_run_list):
    if era == 'HESS1':
        off_runs_filtered = [x for x in off_run_list if x <= 85399]
    if era == 'HESS2':
        off_runs_filtered = [x for x in off_run_list if x > 85399 and x <=123799]
    if era == 'HESS1u':
        off_runs_filtered = [x for x in off_run_list if x > 123799] #and x<=154799]
    return off_runs_filtered
```

```python
def find_bin(distribution, value):
    for i in range(0,len(distribution)-1):
        distribution.sort()
        if value>distribution[i] and value<distribution[i+1]:
            return [distribution[i], distribution[i+1]]
```
