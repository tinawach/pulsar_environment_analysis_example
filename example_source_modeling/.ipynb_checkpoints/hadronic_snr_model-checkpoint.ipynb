# Particle Model to explain the observed Gamma-ray emission

This notebook show the code which was used to compute the physcial model presented in the publication https://arxiv.org/abs/2403.16802.

Using the GAMERA package, a proton population is evolved over the age of the Supernova Remnant and the resulting gamma-ray spectrum compared to the observed gamma-ray emission. 

Then a Markow-Chain-Monte-Carlo optimization is employed to identifiy the confidence intervals of the model parameters best describing the observed emission.

```python
def define_model_parameters_scales(model_parameters):
    for key in model_parameters.keys():
        scale = 100./(model_parameters[key][-1] - model_parameters[key][1])
        model_parameters[key].append(scale) #apply scaling and add to dict
```

```python
model_parameters_snr = {   
    'density_mc' : [60,1,300], #/cm3 #high particle density region
    'alpha_p' : [2.11,1.0,2.59],
    'power_p' : [49.5,47,51], #log10(E/erg)
    'del_alpha' : [-0.72,-1.0,1.0],
    'k_ep' : [-3.2,-7,-1], #ratio of electron spectrum norm to proton log10(Kep)
    'b_now_snr' : [58e-6,3e-7,300e-6], #Gauss
}
define_model_parameters_scales(model_parameters_snr)
```

```python
snr_x = np.concatenate((df_radio['e_ref'], df_fermi_B['e_ref'], df_hess_relic[df_hess_relic['is_ul']==False]['e_ref'][1:]))*gp.TeV_to_erg
snr_y = np.concatenate((df_radio['sed'], np.array(df_fermi_B['sed']), np.array(df_hess_relic[df_hess_relic['is_ul']==False]['sed'][1:])))
snr_yerr = np.concatenate((df_radio['sed_err'], np.array(df_fermi_B['sed_err']), np.array(df_hess_relic[df_hess_relic['is_ul']==False]['sed_err'][1:])))
```

```python
snr_labels = ["alpha_p", "log10(power_p/erg)", "del_alpha", "log10(k_ep)", "b_now_snr (G)"]
snr_fit_parameters = ['alpha_p','power_p','del_alpha','k_ep','b_now_snr']

snr_fit = util.snr_emission(bins, location_snr, model_parameters_snr)
snr_pars = []
snr_lower_bounds = []
snr_upper_bounds = []
for par in snr_fit_parameters:
    snr_pars.append(model_parameters_snr[par][0]*model_parameters_snr[par][-1])
    snr_lower_bounds.append(model_parameters_snr[par][1]*model_parameters_snr[par][-1])
    snr_upper_bounds.append(model_parameters_snr[par][2]*model_parameters_snr[par][-1])
snr_bounds = (snr_lower_bounds, snr_upper_bounds)

mcmc_result_snr = util.run_mcmc(5,snr_pars,snr_fit.log_prob_snr,snr_x,snr_y,snr_yerr)
   
```

```python
mcmc_snr_fit_results, mcmc_snr_fit_results_err_high, mcmc_snr_fit_results_err_low = util.mcmc_results("SNR", snr_labels, snr_fit_parameters, 
                                                                                          model_parameters_snr, mcmc_result_snr)
```

```python
high_results_snr = mcmc_snr_fit_results + mcmc_snr_fit_results_err_high
low_results_snr = mcmc_snr_fit_results - mcmc_snr_fit_results_err_low
```

```python
e_photon = np.logspace(-6,15,bins+20) * gp.eV_to_erg

model_ext = snr_fit.fit_snr_model(e_photon, alpha_p_fit=mcmc_snr_fit_results[0], power_p_fit=mcmc_snr_fit_results[1], del_alpha_fit=mcmc_snr_fit_results[2], k_ep_fit=mcmc_snr_fit_results[3], b_now_snr_fit=mcmc_snr_fit_results[4])
model_ext_high = snr_fit.fit_snr_model(e_photon, alpha_p_fit=high_results_snr[0], power_p_fit=high_results_snr[1], del_alpha_fit=high_results_snr[2], k_ep_fit=high_results_snr[3], b_now_snr_fit=high_results_snr[4])
model_ext_low = snr_fit.fit_snr_model(e_photon, alpha_p_fit=low_results_snr[0], power_p_fit=low_results_snr[1], del_alpha_fit=low_results_snr[2], k_ep_fit=low_results_snr[3], b_now_snr_fit=low_results_snr[4])

f, ax = plt.subplots() 
draw_observations_data(ax)
x = e_photon
ax.plot(x, model_ext, linestyle='--', color='lightsalmon')
ax.plot(x, model_ext_high, linestyle='-', color='lightsalmon', alpha=0.3)
ax.plot(x, model_ext_low, linestyle='-', color='lightsalmon', alpha=0.3)
ax.fill_between(x, model_ext_high, model_ext_low, color='lightsalmon', alpha=0.3)
fig = plt.gcf()
save(fig, 'hadronic_model_mcmc_compB')
```
